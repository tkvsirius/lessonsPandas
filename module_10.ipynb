````python
# Базовые импорты и загрузка данных
import sys
import os
from pathlib import Path
import pandas as pd
import numpy as np

# Настройки отображения pandas
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 120)

DATA_PATH = Path('data') / 'melb_data.csv'
if not DATA_PATH.is_file():
    raise FileNotFoundError(f"Не найден файл: {DATA_PATH.resolve()}")

# Загрузка датасета
melb_data = pd.read_csv(DATA_PATH)

# Быстрый sanity-check
print('Форма набора данных:', melb_data.shape)
print('Первые столбцы:', list(melb_data.columns[:10]))
print('Количество пропусков (топ-5):')
print(melb_data.isna().sum().sort_values(ascending=False).head())
````
