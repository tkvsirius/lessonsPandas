# –û–ø–∏—Å–∞—Ç–µ–ª—å–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Å –º–µ—Ç–æ–¥–æ–º `describe()`

## –û—Å–Ω–æ–≤–Ω–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ

–ú–µ—Ç–æ–¥ `DataFrame.describe()` ‚Äî —ç—Ç–æ –º–æ—â–Ω—ã–π –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ –ø–æ–ª—É—á–µ–Ω–∏—è –æ—Å–Ω–æ–≤–Ω—ã—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –¥–∞–Ω–Ω—ã—Ö. –û—Å–æ–±–µ–Ω–Ω–æ –ø–æ–ª–µ–∑–µ–Ω –Ω–∞ —ç—Ç–∞–ø–µ –ø–µ—Ä–≤–∏—á–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö (EDA - Exploratory Data Analysis).

## –î–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é)

–ü–æ —É–º–æ–ª—á–∞–Ω–∏—é –º–µ—Ç–æ–¥ —Ä–∞–±–æ—Ç–∞–µ—Ç —Å —á–∏—Å–ª–æ–≤—ã–º–∏ —Å—Ç–æ–ª–±—Ü–∞–º–∏ (`int64` –∏ `float64`) –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ª–µ–¥—É—é—â–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏:

| –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å | –û–ø–∏—Å–∞–Ω–∏–µ |
|------------|----------|
| `count` | –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π |
| `mean` | –°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ |
| `std` | –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ |
| `min` | –ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ |
| `25%` | –ü–µ—Ä–≤—ã–π –∫–≤–∞—Ä—Ç–∏–ª—å (25-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å) |
| `50%` | –ú–µ–¥–∏–∞–Ω–∞ (50-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å) |
| `75%` | –¢—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å (75-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å) |
| `max` | –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ |

**üí° –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ:** –ù–∞—É—á–Ω–∞—è –Ω–æ—Ç–∞—Ü–∏—è `9.0e+06` –æ–∑–Ω–∞—á–∞–µ—Ç `9 √ó 10‚Å∂ = 9,000,000`

## –î–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö

–° –ø–æ–º–æ—â—å—é –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ `include=['object']` –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤:

| –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å | –û–ø–∏—Å–∞–Ω–∏–µ |
|------------|----------|
| `count` | –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–ø—É—Å—Ç—ã—Ö —Å—Ç—Ä–æ–∫ |
| `unique` | –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π |
| `top` | –°–∞–º–æ–µ —á–∞—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ (–º–æ–¥–∞) |
| `freq` | –ß–∞—Å—Ç–æ—Ç–∞ –ø–æ—è–≤–ª–µ–Ω–∏—è –º–æ–¥—ã |

## –ü—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è

```python
# –î–ª—è –≤—Å–µ—Ö —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
melb_data.describe()

# –î–ª—è —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
melb_data.describe(include=['object'])

# –î–ª—è –≤—Å–µ—Ö —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö
melb_data.describe(include='all')
```

## –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è

**–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ** –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ –∑–Ω–∞—á–µ–Ω–∏—è –æ—Ç–∫–ª–æ–Ω—è—é—Ç—Å—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ.

**–ö–≤–∞—Ä—Ç–∏–ª–∏** –¥–µ–ª—è—Ç –¥–∞–Ω–Ω—ã–µ –Ω–∞ —á–µ—Ç—ã—Ä–µ —Ä–∞–≤–Ω—ã–µ —á–∞—Å—Ç–∏, –ø–æ–º–æ–≥–∞—è –ø–æ–Ω—è—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π.

````python
# –≤—ã–≤–µ–¥–µ–º –Ω–∞ —ç–∫—Ä–∞–Ω –∑–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ 
# —Ç–æ–ª—å–∫–æ –¥–ª—è —Å—Ç–æ–ª–±—Ü–æ–≤ Distance (—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ—Ç –æ–±—ä–µ–∫—Ç–∞ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏ 
# –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –ú–µ–ª—å–±—É—Ä–Ω–∞), BuildingArea (–ø–ª–æ—â–∞–¥—å –∑–¥–∞–Ω–∏—è) –∏ Price (—Ü–µ–Ω–∞ –æ–±—ä–µ–∫—Ç–∞):
melb_data.describe().loc[:, ['Distance', 'BuildingArea' , 'Price']]
````

````python
#–ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —Ç–∏–ø–∞ –¥–∞–Ω–Ω—ã—Ö object –º–µ—Ç–æ–¥ describe() –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç DataFrame, –≤ –∫–æ—Ç–æ—Ä–æ–º —É–∫–∞–∑–∞–Ω—ã:
#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–ø—É—Å—Ç—ã—Ö —Å—Ç—Ä–æ–∫ (count);
#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π (unique);
#—Å–∞–º–æ–µ —á–∞—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ‚Äî –º–æ–¥–∞ ‚Äî  (top);
#—á–∞—Å—Ç–æ—Ç–∞ ‚Äî –æ–±—ä—ë–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è ‚Äî —ç—Ç–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è (freq) –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ —Ç–∏–ø–∞ object –∏—Å—Ö–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã.

melb_data.describe(include=['object'])
````

````python
#–î–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤ —Å—Ç–æ–ª–±—Ü–µ –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è 
# –∫–∞–∂–¥—ã–π –∏–∑ –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –∑–Ω–∞—á–µ–Ω–∏–π (—Ç.–µ. –Ω–∞–π—Ç–∏ —á–∞—Å—Ç–æ—Ç—É –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞–Ω–∏—è), 
# –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –º–µ—Ç–æ–¥ value_counts().

melb_data['Regionname'].value_counts()
````

````python
#–ß—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥ –±–æ–ª–µ–µ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä—É–µ–º—ã–º –∏ –ø–æ–Ω—è—Ç–Ω—ã–º, 
# –º–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º normalize. –ü—Ä–∏ —É—Å—Ç–∞–Ω–æ–≤–∫–µ –∑–Ω–∞—á–µ–Ω–∏—è 
# —ç—Ç–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –Ω–∞ True —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—Ç—å—Å—è –≤ –≤–∏–¥–µ –¥–æ–ª–∏ 
# (–æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã):
melb_data['Regionname'].value_counts(normalize=True)
````

````python
# –ö–∞–∫–æ–π —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –∏–º–µ–µ—Ç —Å—Ç–æ–ª–±–µ—Ü Coordinates
print(melb_data['Coordinates'].dtype)
````

````python
# –°–∫–æ–ª—å–∫–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ CouncilArea?
print(melb_data['CouncilArea'].isnull().sum())
````

````python
# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π (unique) –≤ —Å—Ç–æ–ª–±—Ü–µ Address –±–æ–ª—å—à–µ, 
# —á–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ Coordinates. 
# –≠—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∞–¥—Ä–µ—Å–∞ –ø–æ–º–µ—á–µ–Ω—ã –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏ 
# –≥–µ–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–º–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏.
print(melb_data['Address'].nunique() > melb_data['Coordinates'].nunique())
````

````python
# –°–∫–æ–ª—å–∫–æ –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤ –æ—Ç –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–æ–º–æ–≤ —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç —Ç–∞—É–Ω—Ö–∞—É—Å—ã 
# (—Ç–∏–ø –æ–±—ä–µ–∫—Ç–∞ ‚Äî t)?
print('–°—Ç–æ–ª–±—Ü—ã:', list(melb_data.columns[:]))

print((melb_data['Type'] == 't').mean() * 100)
````
